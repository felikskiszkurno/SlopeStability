```python
import pygimli as pg
import pygimli.meshtools as mt
```

```python
left = -30
right = 30
depth = 25
```

```python
world = mt.createWorld(start=[left, 0],
               end=[right, -depth],
                       layers=[-5])
pg.show(world);
```

```python
line = mt.createLine(start=[ -30, -20],
               end=[30,-15])
geometry = world + line
pg.show(geometry);
```

```python
body = mt.createPolygon([[-8, -12], [0, -7], [8, -13]], isClosed=True, marker=3, addNodes=5, interpolate="spline")
pg.show(body, showNodes=True);
geometry += body
pg.show(geometry, boundaryMarkers=True); # boundaryMarkers dont work
```

```python
mesh = mt.createMesh(geometry, quality=34, area=1)#, smooth=True)
pg.show(mesh)
mesh.save()
```
