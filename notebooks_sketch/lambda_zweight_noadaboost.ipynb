```python
import matplotlib.pyplot as plt
import pandas as pd
import slopestabilitytools
```

```python
data = pd.read_csv('lambda_zweight_high_contrast_sens_study_RMS.csv')
x = data['Lambda'].to_numpy()
y = data['z_weight'].to_numpy()
data_in = {'pred': data['pred'].to_numpy()}
xi, yi, gridded_data = slopestabilitytools.grid_data(x, y, data_in)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
fig.suptitle('Prediction accuracy depending on lamba and z_weight')
#ax = _ax.flatten()
ax.set_xlabel('Lambda')
ax.set_ylabel('z_weight')
cb = []
im0 = ax.tricontourf(x, y, data['pred'].to_numpy())
cb.append(plt.colorbar(im0, ax=ax, label='Classificaton accuracy - higher=better'))
fig.savefig('high_contrast_sensstudy_RMS.png')
#cb.append(plt.colorbar(im0, ax=ax, label='Prediction accuracy [0-1] - higher=better'))
#fig.savefig('noada_sensstudy_points.png')
```
